```python
import numpy as np
from matplotlib import pyplot as plt

rn = np.random.randn(100)
print(rn)
ys = 200 + rn
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)

plt.title("Sample Visualization")
plt.show()
```
